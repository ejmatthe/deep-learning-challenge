# Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_iter1_df = application_df.drop(columns=["EIN","NAME","ASK_AMT"])
application_iter1_df.head()
```

```python
# Determine the number of unique values in each column.
application_iter1_df.nunique()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ["T9", "T13", "T12", "T2", "T25", "T14", "T29", "T15", "T17"]

# Replace in dataframe
for app in application_types_to_replace:
    application_iter1_df['APPLICATION_TYPE'] = application_iter1_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_iter1_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
threshold = 1000
classifications_to_replace = application_iter1_df["CLASSIFICATION"].value_counts()
classifications_to_replace = classifications_to_replace[classifications_to_replace < threshold].index
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    application_iter1_df['CLASSIFICATION'] = application_iter1_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_iter1_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_iter1_df = pd.get_dummies(application_iter1_df)
application_iter1_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_iter1_df["IS_SUCCESSFUL"]
X = application_iter1_df.drop(columns="IS_SUCCESSFUL")

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Compile, Train and Evaluate the Model Iteration 1

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
iter1_nodes_layer1 =  80
iter1_nodes_layer2 = 30

model_iter1 = tf.keras.models.Sequential()

# First hidden layer
model_iter1.add(tf.keras.layers.Dense(units=iter1_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
model_iter1.add(tf.keras.layers.Dense(units=iter1_nodes_layer2, activation="relu"))

# Output layer
model_iter1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
model_iter1.summary()
```

```python
# Compile the model
model_iter1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Create a callback to save model's weights
checkpoint_path1 = "checkpoints/model_optimized1/model_optimized1.ckpt"

cp_callback1 = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path1,
                                                 save_weights_only=True,
                                                 verbose=1)
```

```python
# Train the model
fit_model = model_iter1.fit(X_train_scaled,y_train,epochs=100, callbacks=[cp_callback1])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model_iter1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
model_iter1.save('AlphabetSoupCharityOptimized1.h5')
```
